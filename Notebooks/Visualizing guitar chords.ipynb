```python
import os
import numpy as np
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
y, sr = librosa.load("C:\\Users\\Rahul\\Desktop\\AI\\Audio\\Classifying music notes\\Guitar_chords\\am\\am12.wav",sr=None)
y1, sr1 = librosa.load("C:\\Users\\Rahul\\Desktop\\AI\\Audio\\Classifying music notes\\Guitar_chords\\a\\a1.wav",sr=None)
```

```python
spec = np.abs(librosa.stft(y, hop_length=512))
plt.figure(figsize=(12, 6))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
plt.title('Spectrogram');
plt.show()


spec = np.abs(librosa.stft(y1, hop_length=512))
plt.figure(figsize=(12, 6))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
plt.title('Spetrogram');

plt.show()
```

```python
spec = librosa.feature.melspectrogram(y=y,sr=sr,n_mels=128 )
plt.figure(figsize=(12, 6))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='linear');
plt.colorbar(format='%+2.0f dB');
plt.title('a ');
plt.show()

spec = librosa.feature.melspectrogram(y=y1,sr=sr1,n_mels=128 )
plt.figure(figsize=(12, 6))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='linear');
plt.colorbar(format='%+2.0f dB');
plt.title('g chord');

plt.show()
```

```python
S = librosa.core.stft(y,hop_length=512)
mels= librosa.feature.melspectrogram(y=y,sr=sr,n_mels=128 )
S_dB = librosa.power_to_db(mels, ref=np.max)
librosa.display.specshow(S_dB, x_axis='time',
y_axis='mel', sr=sr,fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('A2 Mel-frequency spectrogram')
plt.tight_layout()
plt.show()

mels= librosa.feature.melspectrogram(y=y1,sr=sr1,n_mels=128 )
S_dB = librosa.power_to_db(mels, ref=np.max)
librosa.display.specshow(S_dB, x_axis='time',
y_axis='mel', sr=sr,fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('A1# Mel-frequency spectrogram')
plt.tight_layout()
plt.show()
```
